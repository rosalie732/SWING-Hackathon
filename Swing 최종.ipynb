```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Seoul
seoul = pd.read_csv('서울.csv', encoding = "utf-8")
seoul.head()
```

```python
X = seoul.drop(["park"],axis=1)
y = seoul["park"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Linear Regression
lrm = LinearRegression(n_jobs=-1)      # fit_intercept=True
result = lrm.fit(X_train, y_train)
print(result.coef_)
```

```python
forecast = lrm.predict(X_test)
print(forecast)
```

```python
accuracy = lrm.score(X_test, y_test)    # 결정계수
print(accuracy)
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=0.1)  # alpha=0: OLS
```

```python
lasso.fit(X_train, y_train)
print(lasso.intercept_ , lasso.coef_ )
```

```python
forecast = lasso.predict(X_test)
print(forecast[0:10])
```

```python
## training-validation procedure(반복 with train-test-split)
from sklearn.metrics import mean_squared_error

result = np.zeros((10,5))
a = [0.001, 0.002, 0.003, 0.004, 0.005]
for step in range(10):
    X0, X1, y0, y1 = train_test_split(X, y, test_size=0.3)
    for choice in range(len(a)):
        lasso = Lasso(alpha=a[choice])      # sklearn.preprocessing.StandardScaler
        lasso.fit(X0, y0); forecast = lasso.predict(X1)
        result[step,choice] = mean_squared_error(y1,forecast)
print(result)
```

```python
result = pd.DataFrame(result)
result.mean()
```

```python
model = sm.OLS(y, sm.add_constant(X))   # sm.OLS(y,x)와 비교
results = model.fit()   # sm.OLS(y, sm.add_constant(x)).fit()
print(results.summary()) 
```

```python
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import confusion_matrix, mean_squared_error
```

```python
# Random Forest
m_RF = RandomForestClassifier(n_estimators=1000)
m_RF.fit(X_train,y_train)
```

```python
pred_RF = m_RF.predict(X_test)
cf_RF = confusion_matrix(y_test,pred_RF)
print('Random Forest Confusion Matrix')
print(cf_RF)
```

```python
score_RF = m_RF.score(X_test, y_test)
print('Random Forest Accuracy')
print(score_RF)
```

```python
# Boosting
m_boosting = GradientBoostingClassifier()
m_boosting.fit(X_train,y_train)
```

```python
pred_boosting = m_boosting.predict(X_test)
cf_boosting = confusion_matrix(y_test,pred_boosting)
print('Boosting Confusion Matrix')
print(cf_boosting)
```

```python
score_boosting = m_boosting.score(X_test, y_test)
print('Boosting Accuracy')
print(score_boosting)
```

```python
# Bagging
m_bagging = BaggingClassifier()
m_bagging.fit(X_train,y_train)
```

```python
pred_bagging = m_bagging.predict(X_test)
```

```python
cf_bagging=confusion_matrix(y_test,pred_bagging)
print('Bagging Confusion Matrix')
print(cf_bagging)
```

```python
score_bagging=m_bagging.score(X_test, y_test)
print('Bagging Accuracy')
print(score_bagging)
```

```python
# Support Vector Regression
from sklearn.svm import SVR

sv_regressor = SVR(kernel='linear', C=1.0, epsilon=0.1)
sv_regressor.fit(X_train, y_train)
```

```python
y_pred = sv_regressor.predict(X_test)
print(y_pred[:10])
```

```python
from sklearn.metrics import mean_squared_error, explained_variance_score

mse = mean_squared_error(y_test, y_pred)
evs = explained_variance_score(y_test, y_pred)
print("MSE:",mse)
print("EVS:",evs)
```
